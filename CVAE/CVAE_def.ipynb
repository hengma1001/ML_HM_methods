```python
import numpy as np
import torch
from torch import nn, optim
from torch.nn import functional as F
# import torch.utils.data
# from torchvision import datasets, transforms
# from torchvision.utils import save_image

from vae_conv import CVAE 
```

```python
channels = 1
# batch_size = cm_train.shape[0]/100
conv_layers = 4
feature_maps = [64,64,64,64]
filter_shapes = [3,3,3,3]
strides = [1,2,1,1]
dense_layers = 2
dense_neurons = [128, 64]
dense_dropouts = [0, 0]
latent_dim = 3
epochs = 1
image_size = 36
```

```python
autoencoder = CVAE(image_size,channels,conv_layers,feature_maps,
           filter_shapes,strides,dense_layers,dense_neurons,dense_dropouts,latent_dim)
```

```python
input_t = torch.rand(1, 1, 36, 36) 
```

```python
with torch.no_grad(): 
    decoded = autoencoder.decode(autoencoder.reparameterize(*(autoencoder.encode(input_t))))
    embed = autoencoder.reparameterize(*(autoencoder.encode(input_t)))
```

```python
print('original dim', input_t.shape) 
x = input_t
for i, conv_layer in enumerate(autoencoder.encode_conv_layers): 
    x = F.relu(conv_layer(x)) 
    print('encode conv layer', x.shape) 
    
x = x.view(-1, autoencoder.conv_dense_dim) 
print('flatten', x.shape) 

for dense_layer, dropput_layer in zip(autoencoder.encode_dense_layers, autoencoder.encode_dropout_layers): 
    x = F.relu(dense_layer(dropput_layer(x)))    
    print('encode dense layer', x.shape)
mu = autoencoder.latent_mu(x) 
logvar = autoencoder.latent_logvar(x) 
print('mean, logvar', mu.shape, logvar.shape)
```

```python
z = autoencoder.latent_decode(embed)
print('decoded latent layer for dense', z.shape)
for i, dense_layer in enumerate(autoencoder.decode_dense_layers): 
    z = F.relu(dense_layer(z)) 
    print('dense', i,  z.shape)
print('after dense', z.shape)
z = z.view([-1,] + autoencoder.mat_shape) 
print('reshape for conv layer', z.shape)
for i, conv_layer in enumerate(autoencoder.decode_conv_layers[:-1]): 
    z = F.relu(conv_layer(z))        
    print ('reshape for conv layer', i, z.shape) 
z = torch.sigmoid(autoencoder.decode_conv_layers[-1](z))
print('final', z.shape)
```
